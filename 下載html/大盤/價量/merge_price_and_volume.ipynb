```python
import pandas as pd
```

```python
taiex_price = pd.read_csv(r'C:\Users\Atlas\Dropbox\StockLab\下載html\大盤\價\taiex_price_data.csv',
                          index_col='date',parse_dates=True)
taiex_volume = pd.read_csv(r'C:\Users\Atlas\Dropbox\StockLab\下載html\大盤\價量\taiex_price_volume_data.csv',
                        index_col='date',parse_dates=True)
otc_price = pd.read_csv(r'C:\Users\Atlas\Dropbox\StockLab\下載html\大盤\價\otc_price_data.csv',
                        index_col='date',parse_dates=True)
otc_volume = pd.read_csv(r'C:\Users\Atlas\Dropbox\StockLab\下載html\大盤\價量\otc_price_volume_data.csv',
                        index_col='date',parse_dates=True)
```
