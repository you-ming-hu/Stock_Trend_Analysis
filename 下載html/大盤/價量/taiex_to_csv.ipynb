```python
import pathlib
import pandas as pd
from bs4 import BeautifulSoup
```

```python
paths = sorted(list(pathlib.Path(r'D:\股票網頁\股票\價量\上市').iterdir()))
```

```python
table = pd.DataFrame()
for i,p in enumerate(paths):
    date = pd.Timestamp(p.name.replace('.html',''))
    print(f'{date}: {i}/{len(paths)}')
    df = pd.read_html(p)[0]

    assert date.strftime(f'{date.year-1911}年%m月市場成交資訊') == df.columns.levels[0]

    df.columns = df.columns.droplevel(0)
    df[['year','month','day']] = df['日期'].str.split('/',expand=True)
    df.year = (df.year.astype(int)+1911).astype(str)
    df['date'] = pd.to_datetime(df[['year','month','day']])
    df = df.rename({'成交股數':'volume','成交金額':'capital','成交筆數':'count','發行量加權股價指數':'close'},
                   axis=1,errors='raise')
    df = df[['date','close','volume','count','capital']]
    table = table.append(df,ignore_index=True)
```

```python
table.to_csv('taiex_price_volume_data.csv',encoding='utf8',index=False)
```
