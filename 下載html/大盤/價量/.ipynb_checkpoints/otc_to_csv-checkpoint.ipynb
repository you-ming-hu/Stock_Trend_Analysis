```python
import pathlib
import pandas as pd
from bs4 import BeautifulSoup
```

```python
paths = sorted(list(pathlib.Path(r'D:\股票網頁\股票\價量\上櫃').iterdir()))
```

```python
table = pd.DataFrame()
for i,p in enumerate(paths):
    date = pd.Timestamp(p.name.replace('.html',''))
    print(f'{date}: {i}/{len(paths)}')
    bs = BeautifulSoup(p.read_text(encoding='utf8'))
    bs.find('tfoot').decompose()

    df = pd.read_html(bs.prettify())[0]

    assert date.strftime(f'日成交量值指數 資料年月:{date.year-1911}/%m') == df.columns.levels[0][0].replace('\xa0',' ')
    df.columns = df.columns.droplevel(0)

    df[['year','month','day']] = df['日 期'].str.split('/',expand=True)
    df.year = (df.year.astype(int)+1911).astype(str)
    df['date'] = pd.to_datetime(df[['year','month','day']])
    df = df.rename({'成交股數（仟股）':'volume','金額（仟元）':'capital','筆數':'count','櫃買指數':'close'},
                    axis=1,errors='raise')
    df[['volume','capital']] = df[['volume','capital']]*1000
    df = df[['date','close','volume','count','capital']]
    table = table.append(df,ignore_index=True)
```

```python
table.to_csv('otc_price_volume_data.csv',encoding='utf8',index=False)
```
