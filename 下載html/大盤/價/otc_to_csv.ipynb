```python
import pathlib
import pandas as pd
from bs4 import BeautifulSoup
```

```python
paths = sorted(list(pathlib.Path(r'D:\股票網頁\股票\價\上櫃').iterdir()))
```

```python
table = pd.DataFrame()
for i,p in enumerate(paths):
    date = pd.Timestamp(p.name.replace('.html',''))
    print(f'{date}: {i}/{len(paths)}')
    bs = BeautifulSoup(p.read_text(encoding='utf8'))
    bs.find('tfoot').decompose()
    df = pd.read_html(bs.prettify())[0]

    assert date.strftime(f'櫃買指數(月查詢) 資料年月:{date.year-1911}/%m') == df.columns.levels[0][0].replace('\xa0',' ')

    df.columns = df.columns.droplevel(0)

    df['date'] = pd.to_datetime(df['日期'])
    df = df.rename({'開市':'open','最高':'high','最低':'low','收市':'close'},
                    axis=1,errors='raise')
    df = df[['date','open','high','low','close']]
    table = table.append(df,ignore_index=True)
    
```

```python
table.to_csv('otc_price_data.csv',encoding='utf8',index=False)
```
